```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
file_path = 'C:/Users/User/Documents/GitHub/bedConfigurationPhaseDiagram/data/jgrf20757-sup-0003-suplementary.xlsx'
data = pd.read_excel(file_path,header =2,usecols='B:J')
data.columns
```

```python
data.describe()
```

```python
X = data[['Width (m)', 'Depth (m)', 'Slope *1000', 'D50 (mm)', 'U (m/s)', 'R',
       'T (°C)']]
y  = data['Bedform']
```

```python
y.describe()
X.describe()
```

```python

phase_model = DecisionTreeClassifier(random_state=1)
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
phase_model.fit(train_X, train_y)
val_predictions = phase_model.predict(val_X)
```

```python
sum(val_predictions == val_y)/len(val_y)
```

```python
%matplotlib qt
plt.figure()
tree.plot_tree(phase_model, filled=True)
plt.title("Trained Decision Tree")
plt.show()


```
